```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#Reading the csv file as an pandas dataframe (not working on my workplace)
user_review = pd.read_csv(r"C:\Users\matth\Desktop\Master\Database_management\Sales\Final\real_final\real-real-final\Submissions\CSV\UserReviewsClean43LIWC.csv", sep=";", encoding='utf-8')
user_review.head(1)
```

```python
#Drop all duplicates in the data 
user_review = user_review.drop_duplicates()
```

```python
#Selcet all required columns
user_review = user_review[["url", "idvscore", "reviewer", "dateP", "thumbsUp", "thumbsTot"]]
user_review
```

```python
#Turn url into short-url
user_review["Short_url"] = user_review["url"].str.split(".com/").str[-1]
user_review
```

```python
#Rename column names to match the ERD
user_review = user_review.rename(columns={"dateP": "Review_date"})
```

```python
#Strip unnecessary characters from reviewer
user_review["reviewer"] = user_review["reviewer"].str.lstrip("' ")
user_review["reviewer"] = user_review["reviewer"].str.rstrip(" '")
user_review
```

```python
#Strip unnecessary characters from dateP
user_review["Review_date"] = user_review["Review_date"].str.strip("' ")
user_review["Review_date"] = user_review["Review_date"].str.strip(" '")
user_review
```

```python
#Turn dateP into date frame datatype
user_review['Review_date'] = pd.to_datetime(user_review['Review_date'])
type('Review_date')
```

```python
#Trun colmn idvscore into numeric datatype and turn all values that do not meet the critiria into NA
user_review['idvscore'] = pd.to_numeric(user_review['idvscore'], errors='coerce')
user_review['thumbsTot'] = pd.to_numeric(user_review['thumbsTot'], errors='coerce')
user_review['thumbsUp'] = pd.to_numeric(user_review['thumbsUp'], errors='coerce')
user_review
```

```python
user_review['user_review_ID'] = range(1, len(user_review) + 1)
user_review
```

```python
#Replace all NaN to 0
user_review = user_review.fillna(0)
```

```python
#Make new dataframe
user = pd.DataFrame(user_review[['reviewer', 'user_review_ID']])
user
```

```python
# Drop all the duplicate reviewer elements
user = user.drop_duplicates(keep='first')
user
```

```python
# Add a new column 'Reviewer_ID' with incremental integers
user['user_ID'] = range(1, len(user) + 1)
user
```
