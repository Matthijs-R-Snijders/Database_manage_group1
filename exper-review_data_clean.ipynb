```python
#Import libraries
import pandas as pd
import numpy as np
```

```python
#Import the csv file and have this problem

expert_review = pd.read_csv(r"C:\Users\matth\Desktop\Master\Database_management\Sales\Final\real_final\real-real-final\Submissions\CSV\ExpertReviewsClean43LIWC.csv", sep=";", encoding='utf-8')

expert_review.head()
```

```python
# drop duplicate rows to clean table
expert_review = expert_review.drop_duplicates()
```

```python
#drop the unnecessary columns and keep the ones related to the ERD
expert_review = expert_review[["url", "idvscore", "reviewer"]]
expert_review
```

```python
#shorten version of URL in order to create FK
expert_review["short_url"] = expert_review["url"].str.split(".com/").str[-1]

expert_review
```

```python
#delete single quotation to make a clean data 
expert_review["reviewer"]  = expert_review["reviewer"].str.strip("' ")

expert_review["reviewer"]  = expert_review["reviewer"].str.strip(" '")
expert_review
```

```python
#delete double quotation to have a clean data 
expert_review["reviewer"]  = expert_review["reviewer"].str.strip('" ')

expert_review["reviewer"]  = expert_review["reviewer"].str.strip(' "')
expert_review
```

```python
#make a column of Expert Reviewer ID to have a PK
expert_review['expert_reviewer_ID'] = range(1, len(expert_review) + 1)

expert_review
```

```python
# Create a new DataFrame 'expert' containing only the 'reviewer' column from 'expert_review'.
expert = pd.DataFrame(expert_review[['reviewer', 'expert_reviewer_ID']])

expert
```

```python
# Drop all the duplicate reviewer elements
expert = expert.drop_duplicates(keep='first')
expert
```

```python
#make a column of Expert Reviewer ID to have a PK
expert['expert_ID'] = range(1, len(expert) + 1)

expert
```
