```python
#Import libraries
import pandas as pd
import numpy as np
```

```python
#Import the csv file and show the first 5 rows
sales = pd.read_csv(r"C:\Users\matth\Desktop\Master\Database_management\Sales\Final\real_final\real-real-final\Submissions\CSV\sales_raw_data.csv", sep=";", encoding='utf-8')
sales.head()
```

```python
#Drop all duplicates
sales = sales.drop_duplicates()
```

```python
#Select what columns we want
sales = sales[["year", "title", "genre", "international_box_office", "domestic_box_office", "worldwide_box_office", "production_budget", "theatre_count", "avg run per theatre", "url"]]
sales
```

```python
#Create Short_url to use as PK
sales["short_url"] = sales["url"].str.split(".com/").str[-1]
sales
```

```python
#Replace all NaN to 0
sales = sales.fillna(0)
sales
```

```python
#Rename the year and avg run per theatre
sales = sales.rename(columns={"year": "release_year", "avg run per theatre": "Avg_run_per_theatre"})
sales
```

```python
#Replace commas with periods in the selected columns
columns_to_replace = ['international_box_office', 'domestic_box_office', 'worldwide_box_office', 'production_budget', 'Avg_run_per_theatre']
sales[columns_to_replace] = sales[columns_to_replace].apply(lambda x: x.str.replace(',', '.'))
sales
```

```python
#Make a temp dataframe to be able to make the genre table made together with Youssef
T = pd.DataFrame(sales[['genre', 'short_url']])
T
```

```python
#Create genre table
genre = pd.DataFrame(T['genre'])
genre
```

```python
#Keep every first instance of a genre and delete the rest
genre = genre.drop_duplicates(keep='first')
genre
```

```python
#Split the genres and explode
genre['genre'] = genre['genre'].replace('Thriller/Suspense', 'Thriller')
genre['genre'] = genre['genre'].replace('Concert/Performance', 'Concert')
genre
```

```python
#create genre_ID as primary key
genre['genre_ID'] = range(1, len(genre) +1)
genre
```

```python
#Merge the temp dataframe with the genre datagrame to make the has_genre dataframe
df2 = pd.DataFrame(T)
df2 = pd.DataFrame(genre)
has_genre = pd.merge(T, genre, on='genre', how='inner')
has_genre
```

```python
#Remove the genre column
has_genre = has_genre.drop(columns=['genre'])
has_genre
```

```python
#Add the dummy PK called has_genre_ID
has_genre['has_genre_ID'] = range(1, len(has_genre) + 1)
has_genre
```

```python
#Temp file to be able to create subset of data from another DataFrame called has_genre
temp = pd.DataFrame(has_genre[['genre_ID', 'short_url']])
temp
```

```python
#Merge the two dataframes on short_url
df2 = pd.DataFrame(temp)

df2 = pd.DataFrame(sales)

sales = pd.merge(temp, sales, on='short_url', how='inner')

sales
```

```python
#Replace all NaN to 0
sales = sales.fillna(0)
sales
```

```python
#Remove the genre column
sales = sales.drop(columns=['genre'])
sales
```
