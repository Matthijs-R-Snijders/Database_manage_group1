```python
#Import libraries
import pandas as pd
import numpy as np
import dateutil.parser as parser
```

```python
#Read the CSV file and load its constents into the DataFrame
meta = pd.read_csv(r"C:\Users\matth\Desktop\Master\Database_management\Sales\Final\real_final\real-real-final\Submissions\CSV\meta_raw.csv", sep=";", encoding='utf-8')
meta.head()
```

```python
#Drop all duplicates from the CSV file
meta = meta.drop_duplicates()
```

```python
#Select which columns we want
meta = meta[["url", "title", "rating", "runtime", "cast", "director", "RelDate"]]
meta
```

```python
#dateutil.parser library for flexible date_time interpretation

#Functie to change date in different format
def parse_date(date_str):
    try:
        return parser.parse(date_str)
    except:
        return None

# Convert RelDate column to datetime and inserting parse_date-function
meta['RelDate'] = meta['RelDate'].apply(parse_date)
meta 
```

```python
# Convert the 'RelDate' column to datetime objects
meta['RelDate'] = pd.to_datetime(meta['RelDate'])

type('RelDate')
```

```python
#If NaN, than make it 0 in every column
meta  = meta.fillna(0)
meta 
```

```python
# Create a new DataFrame 'director' containing only the 'director' column from 'meta' file
#Used for juntion table
director = pd.DataFrame(meta['director'])
```

```python
#Duplicate rows 
director = director.drop_duplicates(keep='first')
director
```

```python
#Generates IDs from 1 to the number of rows in the DataFrame  
#Add a new column 'Unique_ID' with incremental integers
director['director_ID'] = range(1, len(director) + 1)
director
```

```python
#Merge the data frames director and meta on director
df1 = pd.DataFrame(director)
df2 = pd.DataFrame(meta)

meta = pd.merge(df1, df2, on='director', how='inner')
meta
```

```python
#Remove the director column
meta = meta.drop(columns=['director'])
meta
```
